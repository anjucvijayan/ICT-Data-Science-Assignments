# bar graph of male vs female buyers participated in the sales

```python
data1= data["Buyer Gender"].value_counts()
data1
```

```python
data["Buyer Gender"].value_counts().plot(kind= 'bar')
plt.title("bar graph of male vs female buyers",fontsize= 15)
plt.xlabel("Gender",fontsize=10)
plt.ylabel("count of buyer gender", fontsize = 10)
```
