# 3.Check whether all the zones generate the same amount of sales

```python
# H0 : all zones gererate same amount of sales
# H1 : at least one zone have different amount of sales
```

```python
import scipy.stats as stats
fvalue, pvalue = stats.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
print('f value : ', fvalue, 'p value : ',pvalue)
```

```python
if pvalue <= 0.05:
    print("reject the null hypothesis")
else:
    print("fail to reject null hypothesis")
```
